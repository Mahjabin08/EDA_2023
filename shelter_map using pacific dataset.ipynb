# shelter shape file


generating a random shelter file with latitude and longitude from the dataset file to check the feasibility of adding a shape file on the map with for evacuation purpose, as it is one of key part of the project.



```python
# Create a GeoDataFrame
shelters = gpd.GeoDataFrame()
```

```python
num_shelters = 100 
min_lat, max_lat = 20, 26  # Latitude range
min_lon, max_lon = 88, 92  # Longitude range
```

```python
shelters['Lat'] = [random.uniform(min_lat, max_lat) for _ in range(num_shelters)]
shelters['Long'] = [random.uniform(min_lon, max_lon) for _ in range(num_shelters)]
```

```python
shelters.drop(columns=['Latitude','Longitude'],inplace=True)
```

```python
shelters.head()
```

```python
# Create random shelter names
shelter_names = ['Shelter ' + str(i) for i in range(1, num_shelters + 1)]
shelters['Shelter_Name'] = shelter_names
```

```python
shelters.head()
```

```python
shelters['geometry'] = [Point(lon, lat) for lon, lat in zip(shelters['Long'], shelters['Lat'])]
shelters.set_geometry('geometry', inplace=True)
```

```python
shelters.to_file('shelters.shp')
```

```python
import folium

# Create a map centered around the latitude and longitude of pacific dataset
m = folium.Map(location=[shelters['Lat'].mean(), shelters['Long'].mean()], zoom_start=8)

# Add markers for each shelter
for row in shelters.itertuples():
    folium.Marker([row.Lat, row.Long], tooltip=row.Shelter_Name).add_to(m)

# Display the map
m.save('shelters_map.html')
```
